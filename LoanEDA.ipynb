```python
import pandas as pd
import numpy as np
DATA = './data'
```

```python
df = pd.read_csv(f'{DATA}/Columbia_EDA.csv')

# Data cleaning: remove/replace all remaining null values
df.drop(['desc', 'mths_since_last_delinq', 'id', 'member_id'], axis=1, inplace=True)
df['last_pymnt_d'].fillna(value=0, inplace=True)
df.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# sns.scatterplot(x='annual_inc', y='CLI', hue='grade', data=df)
sns.scatterplot(x='annual_inc', y='CLI', hue='loan_status', data=df)
```

```python
sns.boxplot(x='grade', y='CLI', data=df)
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(x='sub_grade', y='CLI', data=df)
```

```python
grades = df['grade']
grades = list(map(lambda g: ord(g) - 65, grades))
sns.distplot(grades, kde=False)
# sns.distplot(df['CLI'], kde=False)
```

```python
sns.violinplot(x='grade', y='CLI', data=df)
```

```python
sns.catplot(
    x='home_ownership',
    y='CLI',
    hue='grade',
    col='term',
    kind='bar',
    data=df,
    palette='rainbow',
)
```

```python
small_df = df[['CLI', 'loan_status', 'annual_inc', 'int_rate', 'term']]
sns.pairplot(small_df, hue='term')
```

```python
sns.heatmap(small_df.corr(), fmt='.4f')
```

```python
plt.figure(figsize=(20, 20))
sns.heatmap(df.corr(), annot=True, fmt='.4f')
```
